## **Step 1**

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

## **Step 2**
Make some data to plot

````python
x = np.linspace(-2, 2, 100)
y1 = [xi**2 for xi in x]
y2 = [abs(xi) for xi in x]
y3 = [(1/3)*xi**3 for xi in x]
````

## **Step 3**
Plot all 3 curves on one figure

````python
plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
````
